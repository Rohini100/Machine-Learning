```python
#Predict maximum temperature when the minimum temperatures are given
```

```python
#Let us take X as the independent Feature/Variable. here X-> Minimum Temperature(given)
```

```python
X=df[['MinTemp']] 
```

```python
#Let us take y as the dependent/target variable. here y-> Maximum temperature(to be predicted)
```

```python
y=df[['MaxTemp']]
```

```python
#Lets first train_test_split our data
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Split the data set in 70:30 -> Train:Test (ideal)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,test_size=0.3,random_state=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
#Importing the LinearRegression model 
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#first create object class of the LinearRegression Model
```

```python
model=LinearRegression()
```

```python
#Lets now train our ML algorithm -.fit()
```

```python
model.fit(X_train,y_train)
```

```python
#Predict the data on test data -> these are our predicted values

y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
#Accuracy of the test model which is equivalent to r2_score
```

```python
model.score(X_test,y_test)*100
```

```python
#Accuracy pf the train model
```

```python
model.score(X_train,y_train)*100
```

```python
#The accuracy of the test and train models must be nearly equal to each other. 
#Then this can be considered to be good trained/good fit model. Our model tes and train scores are equivalent to eaach other so it's a good fit model.
```

```python
#lets calculate the slope and intercept of our LinearRegression model
model.coef_
```

```python
model.intercept_
```

```python
#Lets calculate r2 Score
```

```python
from sklearn.metrics import r2_score
```

```python
r2=r2_score(y_test,y_pred)
```

```python
r2*100
```

```python
#Adjusted R2 score -> calculated by dividing the residual mean square error by the total mean square error and the result is subtracted from 1
#Adjusted R2 score must be always less than r2 Score
```

```python
n=len(df)
p=len(y_test)
```

```python
Adjusted_r2=1-(1-r2)*(n-1)/(n-p-1)
```
